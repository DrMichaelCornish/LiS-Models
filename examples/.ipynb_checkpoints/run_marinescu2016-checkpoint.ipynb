```python
import os
os.chdir('../') # insures use of models folder as python module 
# Any figures saved or data pickled in the current working 
# directly will be saved up one folder from here.
import pybamm
import models
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
```

```python
model = models.Marinescu2016()
```

```python
model.print_citations()
```

```python
experiment = pybamm.Experiment(['Discharge at 1.7 A until 2.24V'])

sim = pybamm.Simulation(
    model=model,
    experiment = experiment,
    solver=pybamm.CasadiSolver(
        atol=1e-12,
        rtol=1e-12,
        max_step_decrease_count = 15,
    ),
)
```

```python
sim.solve()
```

```python
# See figure 3(a) of Marinescu et al. (2016)
capacity = sim.solution['Capacity [Ah]'].entries
V = sim.solution["Terminal voltage [V]"].entries
plt.plot(capacity, V)

plt.legend(["0.5C",])
plt.xlabel('Capacity [Ah]')
plt.ylabel('Voltage [V]');
```
